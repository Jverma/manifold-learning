```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn import manifold
```


# Swiss Roll dimensional reduction with manifold learning methods

### Plot the original data

```python
X, color = datasets.samples_generator.make_swiss_roll(n_samples=1500)

fig = plt.figure(figsize=[10,10])

ax = fig.add_subplot(211, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
plt.show()
```

## Linear Embedding methods 

Start with the linear dimensional reduction methods to establish the baseline. 

### Principal Component Analysis

```python
X_r = PCA(n_components=2).fit_transform(X)
print(X_r.shape)

fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('PCA')
plt.show()
```

### Multi-Dimensional Scaling

```python
X_r = manifold.MDS(n_components=2, max_iter=100, n_init=1).fit_transform(X)
print(X_r.shape)

fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('MDS')
plt.show()
```

## Non-linear manifold learning methods

### Isomap

```python
X_r = manifold.Isomap(n_neighbors=12, n_components=2).fit_transform(X)
print(X_r.shape)

fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('Isomap')
plt.show()
```

### Locally Linear Embedding

```python
X_r, err = manifold.locally_linear_embedding(X, n_neighbors=12,
                                             n_components=2)
print(X_r.shape)

fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('LLE')
plt.show()
```

### Laplacian Eigenmaps

```python
X_r = manifold.SpectralEmbedding(n_components=2,
                                n_neighbors=12).fit_transform(X)
print(X_r.shape)

fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(212)
ax.scatter(X_r[:, 0], X_r[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('Laplacian Eigenmaps')
plt.show()

```
